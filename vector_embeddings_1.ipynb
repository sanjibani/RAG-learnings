```python
from langchain.document_loaders import PyPDFLoader
```

```python
# Load PDF
loaders = [
    # Duplicate documents on purpose - messy data
    PyPDFLoader("f46d978d-3272-454f-9038-c8753e4aac36.pdf"),
    PyPDFLoader("27d2a29c-811b-4809-bba2-1dad969f0cde.pdf")
]
```

```python
pip install pypdf
```

```python
docs = []
for loader in loaders:
    docs.extend(loader.load())
print(docs)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1500,
    chunk_overlap = 150
)
```

```python
print(text_splitter)
```

```python
splits = text_splitter.split_documents(docs)
```

```python
embedding = OpenAIEmbeddings()
```

```python
sentence1 = "i like dogs"
sentence2 = "i like canines"
sentence3 = "the weather is ugly outside"
```

```python
embeddings1 = embedding.embed_query(sentence1)
embeddings2 = embedding.embed_query(sentence2)
embeddings3 = embedding.embed_query(sentence3)
```

```python
print(embeddings1 , embeddings2 , embeddings3)
```

```python
from langchain.vectorstores import Chroma
```

```python
persist_directory = 'docs/chroma/'
```

```python
!rm -rf ./docs/chroma
```

```python
vectordb = Chroma.from_documents(
    documents=splits,
    embedding=embedding,
    persist_directory=persist_directory
)
```

```python
print(vectordb._collection.count())
```

```python
question = "who're the participants in the earning call"
```

```python
docs = vectordb.similarity_search(question,k=5)
```

```python
len(docs)
```

```python
for doc in docs:
    print(doc.metadata)
```
